```python
import pandas as pd
```

```python
file_path = './log/sim/jobcomp.log'

with open(file_path, 'r', encoding="utf-8") as f:
    lines = f.readlines()

data = {'JobId':[], 'NodeList':[], 'AllocCPUS':[], 'AllocNodes':[], 'Submit':[], 'Start':[], 'End':[]}

index = [[0], [9], [11], [10], [19], [7], [8]]
i = 0
for n in data.keys():
    index[i].append(n)
    i += 1
```

```python
for line in lines:
    line = line.split(' ')
    for i in index:
        data[i[1]].append(line[i[0]].split('=')[1])
```
