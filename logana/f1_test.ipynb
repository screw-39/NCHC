```python
waiting_time = pd.read_parquet('working_time.parquet')
average_time = waiting_time.groupby('#CPU').describe().iloc[:,1]
#waiting_time = waiting_time.sort_values(by=['#CPU'])
```

```python
average_time[average_time > 1]
```

```python
fig = go.Figure(go.Scatter(
    x=waiting_time['#CPU'],
    y=waiting_time['Working Time'],
    name='working time',
    mode='markers',
    marker={'opacity':0.4}
    ))

fig.update_traces(marker_color='rgb(0, 0, 0)', marker_line_color='rgb(0, 0, 0)',
                    marker_line_width=1.5, opacity=0.3)

fig.add_trace(go.Scatter(x=average_time[average_time > 1].index, y=average_time[average_time > 1], name='average'))

fig.update_layout(
    plot_bgcolor='White',
    yaxis={
        'type':'log'
    },
    xaxis={
        'title':'#CPU',
        'type':'log',
        'range':[-0.1,4.8]
    },
    title={'text':'#CPU vs working time', 'font':{'size':30}},
    xaxis_title={'text':'#CPU', 'font':{'size': 30}},
    yaxis_title={'text':'Seconds', 'font':{'size':30}},
    height=600,
    width=1700
)

# fig.add_shape(
#         type='line', line=dict(dash='dash'),
#         name = '112',
#         label = {'text':'ct112'},
#         x0 = 112,
#         x1 = 112,
#         y0 = 0.5,
#         y1 = 10000000
#     )

# fig.add_shape(
#         type='line', line=dict(dash='dash'),
#         name = '224',
#         label = {'text':'ct224'},
#         x0 = 224,
#         x1 = 224,
#         y0 = 0.5,
#         y1 = 10000000
#     )

# fig.add_shape(
#         type='line', line=dict(dash='dash'),
#         name = '1120',
#         label = {'text':'ct1k'},
#         x0 = 1120,
#         x1 = 1120,
#         y0 = 0.5,
#         y1 = 10000000
#     )

# fig.add_shape(
#         type='line', line=dict(dash='dash'),
#         name = '2240',
#         label = {'text':'ct2k'},
#         x0 = 2240,
#         x1 = 2240,
#         y0 = 0.5,
#         y1 = 10000000
#     )

# fig.add_shape(
#         type='line', line=dict(dash='dash'),
#         name = '4480',
#         label = {'text':'ct4k'},
#         x0 = 4480,
#         x1 = 4480,
#         y0 = 0.5,
#         y1 = 10000000
#     )

# fig.add_shape(
#         type='line', line=dict(dash='dash'),
#         name = '8960',
#         label = {'text':'ct8k'},
#         x0 = 8960,
#         x1 = 8960,
#         y0 = 0.5,
#         y1 = 10000000
#     )

fig.add_shape(
        type='line', line=dict(dash='dash'),
        name = 'min',
        label = {'text':'min'},
        x0 = 0,
        x1 = 60000,
        y0 = 60,
        y1 = 60
    )

fig.add_shape(
        type='line', line=dict(dash='dash'),
        name = 'hour',
        label = {'text':'hour'},
        x0 = 0,
        x1 = 60000,
        y0 = 3600,
        y1 = 3600
    )

fig.add_shape(
        type='line', line=dict(dash='dash'),
        name = 'day',
        label = {'text':'day'},
        x0 = 0,
        x1 = 60000,
        y0 = 3600*24,
        y1 = 3600*24
    )

fig.add_shape(
        type='line', line=dict(dash='dash'),
        name = 'week',
        label = {'text':'week'},
        x0 = 0,
        x1 = 60000,
        y0 = 3600*24*7,
        y1 = 3600*24*7
    )

fig.show()
```
